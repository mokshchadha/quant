```python
# macd in this we get the moving avg of fast and slow steps (12 and 26 steps) and our signal line is the 9 steps
import yfinance as yf
tickers = ["AMZN", "GOOG", "FB"]

ohlcv_data = {}

for ticker in tickers:
    tempdata = yf.download(ticker, period="1mo", interval="15m", multi_level_index=False)
    tempdata.dropna(inplace=True)
    ohlcv_data[ticker] = tempdata
```

```python
def MACD(DF, a=12, b= 26, c =9):
    df = DF.copy() # we dont want to change the original data while working in prod as it takes tiem to get the data
    df["ma_fast"] = df['Close'].ewm(span = a).mean()
    df['ma_slow'] = df['Close'].ewm(span = b).mean()
    df["macd"] = df['ma_fast'] - df['ma_slow'] # this is the macd that we are after fast - slow
    df['signal'] = df['Close'].ewm(span=c).mean()
    return df["macd"], df['signal']

# MACD(ohlcv_data['FB'])

for ticker in tickers:
     macd, signal = MACD(ohlcv_data[ticker])
     ohlcv_data[ticker]['macd'] = macd
     ohlcv_data[ticker]['signal'] = signal


print(ohlcv_data)
```
